```python
from nltk.tokenize import word_tokenize

from textattack.shared.data import NAMED_ENTITIES
from textattack.transformations import WordSwapExtend
from textattack.transformations import WordSwapWordNet
import pandas as pd
from random import randint
import random
from transformers import pipeline
import datasets
from datasets import load_dataset
from textattack.augmentation import CheckListAugmenter
from textattack.augmentation import CLAREAugmenter
from textattack.transformations import  WordSwapChangeName
from textattack.transformations import WordSwapRandomCharacterDeletion
from textattack.transformations import WordSwapChangeLocation
from textattack.transformations import WordSwapExtend
from textattack.transformations import WordSwapWordNet


from textattack.transformations import CompositeTransformation
from textattack.augmentation import Augmenter
from numpy import random
import numpy as np
from textattack.transformations import WordSwap



data=pd.read_csv("/content/data.csv")



# Vegetables and Fruits dictionary

vegetables_and_fruits=data["vegetables and fruits"]
vegetables_and_fruits=np.array(vegetables_and_fruits)
vegetables_and_fruits=vegetables_and_fruits[vegetables_and_fruits!="1"]
vegetables_and_fruits=[vf.lower() for vf in vegetables_and_fruits]

plural_vegetables_and_fruits=[vg+"s" for vg in vegetables_and_fruits]

# Buildings dictionary

buildings=np.array(data["buildings"])
buildings=np.array(buildings)
buildings=buildings[buildings!="1"]
buildings=[building.lower() for building in buildings]

plural_buildings=[building+"s" for building in buildings]


# Animals dictionary

animals=data["animals"]
animals=np.array(animals)
animals=animals[animals!="1"]
animals=[animal.lower() for animal in animals]

plural_animals=[animal+"s" for animal in animals]



# Printable  dictionary

printables=data["printable"]
printables=np.array(printables)
printables=printables[printables!="1"]
printables=[printable.lower() for printable in printables]

plural_printables=[printable+"s" for printable in printables]


# Sports dictionary

sports=data["sports"]
sports=np.array(sports)
sports=sports[sports!="1"]
sports=[sport.lower() for sport in sports]


# Colors dictionary

colors=data["colors"]
colors=np.array(colors)
colors=colors[colors!="1"]
colors=[str(color).lower() for color in colors]


# Technology  dictionary
technology=data["technology"]
technology=np.array(technology)
technology=technology[technology!="1"]
technology=[t.lower() for t in technology]
plural_technologies=[t+"s" for t in technology]


# Awards dictionary

awards=data["Awards"]
awards=np.array(awards)
awards=awards[awards!="1"]
awards=[award.lower() for award in awards]

plural_awards=[award+"s" for award in awards]

# Professions dictionary

professions=data["professions"]
professions=np.array(professions)
professions=professions[professions!="1"]
professions=[profession.lower() for profession in professions]


plural_prof=[profession+"s" for profession in professions]

# Nationalities


nationalities=data["nationalities"]
nationalities=np.array(nationalities)
nationalities=nationalities[nationalities!="1"]
nationalities=[nationality.lower() for nationality in nationalities]

plural_nationalities=[nationality+"s" for nationality in nationalities]


# Numerical rankings dictionary
numerical_rankings=data["numerical rankings"]
numerical_rankings=np.array(numerical_rankings)
numerical_rankings=numerical_rankings[numerical_rankings!="1"]
numerical_rankings=[nr.lower() for nr in numerical_rankings]

#Months dictionary

months=["january","february","march","april","may","june","july","august","september","october","november","december"]


def check_first_award(sentence):
  index=-1
  award_found=""
  for award in awards:
    if sentence.find(award)!=-1:
      index=sentence.index(award)
      award_found=award
      break
  return index, len(award_found)

def change_award(sentence):
  sentence=sentence.lower()
  indx,ln_award=check_first_award(sentence)
  if indx!=-1:
    index=random.randint(len(awards))
    new_sentence=sentence[:indx]+awards[index]+sentence[indx+ln_award+1:]
    return new_sentence
  else:
    return sentence



```

```python
def sentence_with_answer(context,answer_start):
  if len(context)==1:
    return context[0]
  if len(context)==0:
    return ""
  for i in range(100000000):
    if (answer_start-i)<0:
      start_=0
      break

    if context[answer_start-i]==".":
      if context[answer_start-i+1]==" ":
        start_=answer_start-i+2
      else:
        start_=answer_start-i+1
      break

  for i in range(100000000):
    if (answer_start+i)>=len(context):
      end_=len(context)-1
      break
    if context[answer_start+i]==".":
      end_=i+answer_start+1
      break
  return context[start_:end_]


```

```python


def return_capital(word,gen_word):
  if word[0].isupper():
        gen_word=gen_word[0].upper()+gen_word[1:]
  return gen_word




def augmentation_by_sentence(sentence):
  #sentence=change_award(sentence)
  list_words=word_tokenize(sentence)
  new_text=""

  for word in list_words:
    gen_list=augmenter.augment(word)
    #AUGMENTATION VEGETABLES AND FRUITS
    if word.lower() in vegetables_and_fruits:
      index=random.randint(len(vegetables_and_fruits))
      gen_list[0]=vegetables_and_fruits[index]
    if word.lower() in plural_vegetables_and_fruits:
      index=random.randint(len(plural_vegetables_and_fruits))
      gen_list[0]=plural_vegetables_and_fruits[index]
    #AUGMENTATION BUILDINGS
    if word.lower() in buildings:
      index=random.randint(len(buildings))
      gen_list[0]=buildings[index]
    if word.lower() in plural_buildings:
      index=random.randint(len(plural_buildings))
      gen_list[0]=plural_buildings[index]
     #AUGMENTATION ANIMALS
    if word.lower() in animals:
      index=random.randint(len(animals))
      gen_list[0]=animals[index]
    if word.lower() in plural_animals:
      index=random.randint(len(plural_animals))
      gen_list[0]=plural_animals[index]
    #AUGMENTATION PRINTABLES
    if word.lower() in printables:
      index=random.randint(len(printables))
      gen_list[0]=printables[index]
    if word.lower() in plural_printables:
      index=random.randint(len(plural_printables))
      gen_list[0]=plural_printables[index]
    #AUGMENTATION SPORTS
    if word.lower() in sports:
      index=random.randint(len(sports))
      gen_list[0]=sports[index]
     #AUGMENTATION COLORS
    if word.lower() in colors:
      index=random.randint(len(colors))
      gen_list[0]=colors[index]
     #AUGMENTATION TECHNOLOGY
    if word.lower() in technology:
      index=random.randint(len(technology))
      gen_list[0]=technology[index]
    if word.lower() in plural_technologies:
      index=random.randint(len(plural_technologies))
      gen_list[0]=plural_technologies[index]
    #AUGMENTATION AWARDS
    if word.lower() in awards:
      index=random.randint(len(awards))
      gen_list[0]=awards[index]
    if word.lower() in plural_awards:
      index=random.randint(len(plural_awards))
      gen_list[0]=plural_awards[index]
    #AUGMENTATION PROFESSIONS
    if word.lower() in professions:
      index=random.randint(len(professions))
      gen_list[0]=professions[index]
    if word.lower() in plural_prof:
      index=random.randint(len(plural_prof))
      gen_list[0]=plural_prof[index]
     #AUGMENTATION NATIONALITIES
    if word.lower() in nationalities:
      index=random.randint(len(nationalities))
      gen_list[0]=nationalities[index]
    if word.lower() in plural_nationalities:
      index=random.randint(len(plural_nationalities))
      gen_list[0]=plural_nationalities[index]
    #AUGMENTATION NUMERICAL RANKING
    if word.lower() in numerical_rankings:
      index=random.randint(len(numerical_rankings))
      gen_list[0]=numerical_rankings[index]
    #AUGMENTATION MONTHS
    if word.lower() in months:
      index=random.randint(len(months))
      gen_list[0]=months[index]
    #AUGMENTATION LOCATION
    if word in NAMED_ENTITIES["country"] :
      index=random.randint(len(NAMED_ENTITIES["country"]))
      gen_list[0]=NAMED_ENTITIES["country"][index]
    if word in NAMED_ENTITIES["city"] :
      index=random.randint(len(NAMED_ENTITIES["city"]))
      gen_list[0]=NAMED_ENTITIES["city"][index]
    #AUGMENTATION NUMERICAL RANKING
    if word.lower() in numerical_rankings:
      index=random.randint(len(numerical_rankings))
      gen_list[0]=numerical_rankings[index]
    gen_list[0]=return_capital(word,gen_list[0])
    if len(new_text)==0:
      new_text=new_text+gen_list[0]
    else:
      new_text=new_text+" "+ gen_list[0]
  return new_text



```

```python
def add_augmented_sentence(text,gtext,ans_start):
  x = random.randint(2)
  if x==0:
    if text[-1]==".":
      text=text+" "+gtext
    else:
      text=text+". "+gtext
    ans_start=ans_start
  if x==1:
    if text[-1]==".":
      text=gtext+" "+text
      ans_start=len(gtext)+1+ans_start
    else:
      text=gtext+". "+text
      ans_start=len(gtext)+2+ans_start
  return ans_start,text

proof_sentence="Apple Oranges House Schools Cat Dogs Magazine Books Basketball."
prof_augmented=augmentation_by_sentence(proof_sentence)
a_start,augsnt=add_augmented_sentence(proof_sentence,prof_augmented,20)
print(augsnt)
print(a_start)
print(augsnt[a_start:a_start+7])
```

```python
#data_all["answers"]["text"]
all_ids=[]
def augmentation_data(data_all):

  rand=random.uniform(0, 1)
  if rand>=0.75:
    contexts=data_all["context"]
    questions=data_all["question"]
    answers=data_all["answers"]
    ids=data_all["id"]
    titles=data_all["title"]

    augmented_sntcs=[]
    aug_answers=[]
    aug_ids=[]
    aug_questions=[]
    aug_titles=[]
    for i in range(len(contexts)):
      context=contexts[i]
      #print(context)
      ans_start=answers[i]["answer_start"][0]
      txt_ans=answers[i]["text"][0]
      snt_ans=sentence_with_answer(context,ans_start)
      aug_snt=augmentation_by_sentence(snt_ans)
      #print(aug_snt)
      new_ans_start,new_aug_snt=add_augmented_sentence(context,aug_snt,ans_start)
      #print(new_aug_snt)

      aug_answers.append({'text': [txt_ans],'answer_start': [new_ans_start]})
      augmented_sntcs.append(new_aug_snt)
      nw_id=""
      for letter in ids[i]:
        nw_id=letter+nw_id

      aug_ids.append(nw_id)
      aug_questions.append(questions[i])

      aug_titles.append(titles[i])


    data_all["context"]=data_all["context"]+augmented_sntcs
    data_all["answers"]=data_all["answers"]+aug_answers
    data_all["id"]=data_all["id"]+aug_ids
    data_all["question"]=data_all["question"]+aug_questions
    data_all["title"]=data_all["title"]+aug_titles

  return data_all
```

```python
augmented_dataset=dataset.map(augmentation_data, batched=True, batch_size=10)
```
